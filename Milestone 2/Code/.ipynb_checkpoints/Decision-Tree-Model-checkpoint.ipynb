```python
def training_testing_split(data_dec_tree, size):
    #takes in a proportion of rows to split or a number of rows to split.
    if isinstance(size, float):
        size = round(size*len(data_dec_tree))#This allows the function to work with proportions
    
    indicies = data_dec_tree.index.tolist()
    testindicies = random.sample(population=indicies, k=size)#pulls random rows from data

    test_data = data_dec_tree.loc[testindicies]#set testing data to the indices selected
    train_data = data_dec_tree.drop(testindicies)#set training to everything else

    return train_data,test_data

def check_end(data):
    
    outcome_column = data[:, -1]
    unique_classes = np.unique(outcome_column)

    if len(unique_classes) == 1:
        return True
    else:
        return False
```

```python
#tree = dec_tree_gini
tree_data = training_base.values
tree_data[:5]
print(tree_data[:5])

def classify_data(data):
    #this can be considered a "leaf" of the tree, or an end.
    labels = data[:, -1]#array of outcomes
    unique_classes, counts_unique_classes = np.unique(labels, return_counts = True)
    
    
    index = counts_unique_classes.argmax()
    classification = unique_classes[index]
    
    return classification

```

```python

def get_pot_splits(data):
    #get potential places to split, find all unique values in dataset and
    #add to potential split vals when not zero.
    pot_splits = {}
    _, n_cols = data.shape
    for col in range(n_cols - 1):        # excluding the last column which is the label
        pot_splits[col] = []
        vals = data[:, col]
        unique_vals = np.unique(vals)

        for val in range(len(unique_vals)):
            if val != 0:
                currval = unique_vals[val]#new tail
                preval = unique_vals[val - 1] #previous tail of array
                split = (currval + preval) / 2 #midpoint of the values (split in half)
            
                pot_splits[col].append(split)
    
    return pot_splits

```

```python

def split_data(data, split_column, split_value):
    #simple splitting function
    split_column_values = data[:, split_column]

    data_below = data[split_column_values <= split_value]
    data_above = data[split_column_values >  split_value]
    
    return data_below, data_above



#print(classification)
#print(labels)
```

```python
#caclulating entropy helper functions



def calc_indv_entropy(data):
    #caclulating entropy for entropy for a subset of data
    label_column = data[:, -1]
    _, counts = np.unique(label_column, return_counts=True)

    probabilities = counts / counts.sum()
    entropy = sum(probabilities * -np.log2(probabilities))#log base 2 of probabilities
     
    return entropy

def calc_entropy(data_below, data_above):
    #calculate entropy split, each entropy determined by the size of each part.
    n = len(data_below) + len(data_above)
    p_data_below = len(data_below) / n#proportion for lower, upper half
    p_data_above = len(data_above) / n

    overall_entropy =  (p_data_below * calc_indv_entropy(data_below) 
                      + p_data_above * calc_indv_entropy(data_above))# simple summation
    
    return overall_entropy

def find_best_split(data, pot_splits):
    entropy = 12345# set initial entropy very high
   # best_split_column = 5
   # best_split_value = pot_splits[5]
    
    for i in pot_splits:
        for val in pot_splits[i]:
            data_below, data_above = split_data(data, split_column=i, split_value=val)
            current_entropy = calc_entropy(data_below, data_above)
            if current_entropy <= entropy:
                entropy = current_entropy
                best_split_column = i
                best_split_value = val
    
    return best_split_column, best_split_value# return column and value of the best place to split the database.

    

```

```python

#Decision_tree using entropy
def decision_tree(data, counter=0):
    # preparing data
    if counter == 0:
        dataVals = data.values
    else:
        dataVals = data           
    
    # base case: Act to break the recusive nature of tree, done via check_purity
    #when it is pure it classifies the data
    if check_end(dataVals):
        classification = classify_data(dataVals)
        return classification

    
    # recursive part
    else:    
        counter += 1#increment counter

        # calling the helper functions to get the split value
        potential_splits = get_pot_splits(dataVals)
        split_column, split_value = find_best_split(dataVals, potential_splits)
        data_below, data_above = split_data(dataVals, split_column, split_value)
        
        # instantiate sub-tree, visually formatted
        question = "{} <= {}".format(split_column, split_value)
        sub_tree = {question: []}#Creating a subtree to contain a set of paths
        
        # find whether diabetes is detected or not
        yes_found = decision_tree(data_below, counter)
        no_found = decision_tree(data_above, counter)
        
        if yes_found == no_found:#handle case they are the same
            sub_tree = yes_found
        else:
            sub_tree[question].append(yes_found)
            sub_tree[question].append(no_found)
        
        return sub_tree
```

```python
def classify_example(example, tree):
    #clasify a given entry via the decision tree made.
    paths = list(tree.keys())[0]
    feature_name, comparison_operator, value = paths.split()

    # does it fit this path?
    if example[feature_name] <= float(value):
        selection = tree[paths][0]
    else:
        selection = tree[paths][1]
        
    else: #covers when feature is categorical
        if str(example[feature_name]) == value:
            selection = tree[paths][0]
        else:
            selection = tree[paths][1]

    # base case(at a leaf)
    if not isinstance(selection, dict):
        return selection 
    
    # recursive part(travel down tree)
    else:
        residual_tree = selection
        return classify_example(example, residual_tree)
```

```python
def calculate_accuracy(data, tree):
        #simple calculation for the accuracy of the decision tree,
        # we classify each example in the tree via mapping and then find the differences.
    data["classification"] = data.apply(classify_example, axis=1, args=(tree,))
    data["classification_correct"] = data["classification"] == data["label"]
    
    accuracy = data["classification_correct"].mean()
    
    return accuracy
acc = calculate accuracy(testing_base, tree)
```

```python
#decision tree for random forest, includes random_subspace parameter.
def decision_tree_rand_forest(data, counter=0, random_subspace):
    # preparing data
    if counter == 0:
        dataVals = data.values
    else:
        dataVals = data           
    
    # base case: Act to break the recusive nature of tree, done via check_purity
    #when it is pure it classifies the data
    if check_end(dataVals):
        classification = classify_data(dataVals)
        return classification

    
    # recursive part
    else:    
        counter += 1#increment counter

        # calling the helper functions to get the split value
        potential_splits = get_pot_splits_rand_forest(dataVals,random_subspace)
        split_column, split_value = find_best_split(dataVals, potential_splits)
        data_below, data_above = split_data(dataVals, split_column, split_value)
        
        # instantiate sub-tree, visually formatted
        question = "{} <= {}".format(split_column, split_value)
        sub_tree = {question: []}#Creating a subtree to contain a set of paths
        
        # find whether diabetes is detected or not
        yes_found = decision_tree_rand_forest(data_below, counter,random_subspace)
        no_found = decision_tree_rand_forest(data_above, counter,random_subspace)
        
        if yes_found == no_found:#handle case they are the same
            sub_tree = yes_found
        else:
            sub_tree[question].append(yes_found)
            sub_tree[question].append(no_found)
        
        return sub_tree
```

```python
def get_pot_splits_rand_forest(data, random_subspace):
    #get potential places to split, find all unique values in dataset and
    #add to potential split vals when not zero.
    pot_splits = {}
    _, n_cols = data.shape
    
    indices = list(range(n_cols-1))
    #randomly sample columns based off of the random subspace.
    if random_subspace and random_subspace <= len(indices):
        indices = random.simple(population=indices,k=random_subspace)
    
    for col in range(n_cols - 1):        # excluding the last column which is the label
        pot_splits[col] = []
        vals = data[:, col]
        unique_vals = np.unique(vals)

        pot_splits[col] = unique_vals
    
    return pot_splits

```
